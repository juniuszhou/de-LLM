simple linear model demo

```python
import torch
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(400, 300),
    nn.ReLU(),
    nn.Linear(300, 2)
)

# two sample data
input_tensor = torch.randn(2, 400)
output = model(input_tensor)
print(output)

loss = nn.MSELoss()
target = torch.randn(2, 2)
print(target)

optimizer = torch.optim.SGD(model.parameters(), lr=0.001)


for index in range(50):
    optimizer.zero_grad()
    output = model(input_tensor)
    loss_value = loss(output, target)
    loss_value.backward()
    optimizer.step()
    if index % 10 == 0:
        print(f"Epoch {index}, Loss: {loss_value.item()}")

```
