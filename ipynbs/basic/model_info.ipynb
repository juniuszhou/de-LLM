```python
import torch 
import torch.nn as nn
import torch.nn.functional as F

class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 10)
        self.fc2 = nn.Linear(10, 10)
    def forward(self, x):
        x = self.fc1(x)
        # no parameters for pure functions like softmax and gelu
        x = F.softmax(x, dim=1)
        x = F.gelu(x)
        x = self.fc2(x)
        return x

model = MyModel()

# get model parameters with name
for param in model.named_parameters():
    shape = param[1].shape
    size = 1
    for i in shape:
        size *= i

    print(param[0], param[1].shape, size)
```
