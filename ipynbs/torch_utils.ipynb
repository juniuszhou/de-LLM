Demo to use some functions in torch

```python
import torch
import torch.nn.functional as F
a = torch.tensor([1.0, 2.0, 3.0])
# convert the distribution to a probability distribution, means all items add up to 1
norm = F.softmax(a, dim=0)
print(norm)
```

```python
import numpy as np

class MyLinear:
    def __init__(self, input_size: int):
        self.weight = np.random.randn(input_size)
        self.bias = np.random.randn(1)
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return x @ self.weight + self.bias
    def __call__(self, x: np.ndarray) -> number:
        return self.weight @ x + self.bias

l = MyLinear(10)
x = np.random.randn(10)
print(l(x))
```
